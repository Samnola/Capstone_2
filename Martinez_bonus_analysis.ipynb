#### Filtering, Sorting & Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
customers_df = pd.read_csv('EmporiUm_files/customer_list.csv', sep='|')
products_df = pd.read_csv ('EmporiUm_files/Products.csv')
categories_df = pd.read_csv('EmporiUm_files/ProductCategories.csv')
stores_df = pd.read_csv ('EmporiUm_files/StoreDetail.csv')
sales_df = pd.read_csv('EmporiUm_files/StoreSales.csv')
```

1. Create a list of all transactions from January 2024, sorted by sale amount from highest to lowest.
Which transactions had a sale amount greater than $500

```python
# merge sales with product data
sales_data = pd.merge(sales_df, products_df, on='Prod Num')

# merge with category data to get category names
sales_data = pd.merge(
    sales_data,
    categories_df,
    on=['CategoryID', 'SubcategoryID']
)
# merge with store data to include location information
sales_data = pd.merge(sales_data, stores_df, on='Store ID')
```

```python
# convert Transaction Date to datetime format
sales_data['Transaction Date'] = pd.to_datetime(sales_data['Transaction Date'])

# filter for January 2024 transactions
jan_2024_data = sales_data[
    (sales_data['Transaction Date'].dt.year == 2024) &
    (sales_data['Transaction Date'].dt.month == 1)
]
# sort transactions by Sale Amount from highest to lowest
jan_2024_sorted = jan_2024_data.sort_values(by='Sale Amount', ascending=False)

# filter for transactions greater than $500
high_value_transactions = jan_2024_sorted[
    jan_2024_sorted['Sale Amount'] > 500
]

# display results
high_value_transactions
high_value_transactions.tail(5)
```

2. Find all products whose product number begins with 10525. What category and subcategory do they
belong to? Products that begin with "10525" belong to the Technology & Accessories category and the Tablets Subcategory. 

```python
# make sure product numbers are treated as text so we can check how they start
sales_data['Prod Num'] = sales_data['Prod Num'].astype(str)

# filter for products where the product number begins with 10525
filtered_products = sales_data[
    sales_data['Prod Num'].str.startswith('10525')
]
# show the product number along with its category 
# drop_duplicates() removes repeats so the output is easier to read
filtered_products[['Prod Num', 'Category', 'Subcategory']].drop_duplicates()
```

3. Which stores in your assigned territories had at least one transaction with a sale amount over
$1,000? How many such transactions did each store have?

```python
# filter for transactions over $1000
high_sales = sales_data[sales_data['Sale Amount'] > 1000]

# limit to your assigned territories (Florida and Texas)
territory_sales = high_sales[
    (high_sales['State'] == 'Florida') |
    (high_sales['State'] == 'Texas')
]

# group by Store ID AND Store Location so we can see names
store_counts = territory_sales.groupby(
    ['Store ID', 'Store Location']
)['Sale Amount'].count()

store_counts
```

#### Aggregation & NumPy

4. What is the total sales revenue across all in-store transactions in the dataset? What is the average
transaction amount? What is the standard deviation?

```python
# calculate total sales revenue by adding all transaction amounts
total_revenue = sales_data['Sale Amount'].sum()

# calculate the average transaction amount
average_amount = sales_data['Sale Amount'].mean()

# calculate the standard deviation to measure how spread out the values are
std_dev = sales_data['Sale Amount'].std()

# display results
print("Total Sales Revenue:", total_revenue)
print("Average Transaction Amount:", average_amount)
print("Standard Deviation:", std_dev)
```
